# 3. Practicing with magnitudes

As we discussed in lecture, magnitudes appear everywhere in astronomy. Let's practice using magnitudes with our Gaia stars. Remember, the definition of magnitude is:
$$ m_1 - m_2 = - 2.5 \log_{10} (F_1 / F_2) $$
where $m_1, m_2$ are the magnitudes of two stars (creatively named 1 and 2) and $F_1, F_2$ are the fluxes of those two stars. 

**Task:** To get some practice, let's convert the magnitudes listed in the table to fluxes. Let's calculate our fluxes relative to the Sun, so we'll set $m_2 = −26.83$ and we'll calculate $F_1 / F_\mathrm{Sun}$, where $F_\mathrm{Sun}$ is the flux (aka. brightness) of the Sun as measured from Earth. This means we want to calculate the ratio of the stars' brightness to the Sun's brightness.

Make a histogram of $\log_{10}(F_1 / F_\mathrm{Sun})$ for the $g$, $r$, and $b$ Gaia filters (the Gaia magnitudes are stored under the $\texttt{gmag, rmag, bmag}$ columns).

Bonus: write a function for the flux to magnitude conversion

```python
def magnitude_to_flux(magnitude):
    return 10**( (magnitude + 26.83) / -2.5 )

for band in 'grb':
    magnitude = gaia[band+'mag']
    flux_relative_to_sun = magnitude_to_flux(magnitude)
    
    plt.hist(np.log10(flux_relative_to_sun))
    plt.xlabel('$\log_{10}'+f'(F_{band}'+'/F_\mathrm{sun})$')
    plt.show()
```

**Task:** Based on your histograms how many orders of magnitude brighter is the Sun than the typical nearby star?

**Example answer:** Most stars within 15 pc are 15 orders of magnitude fainter than the Sun!! That's 0.000000000000001!! The brighest nearby star is 12 orders of magnitude fainter than the Sun.

What else do these histograms teach us? We know the Sun is 1 AU from Earth (that's $4.848 \times 10^{-6}$ pc) and we also know the flux of a star is related to it's luminosity as:
$$  F = \frac{L}{4 \pi d^2} $$
where $d$ is the distance to the star. This means the apparent brightness of a star relative to the apparent brightness of the Sun is:
$$  F / F_\mathrm{sun}= \frac{ L d_\mathrm{Earth~to~Sun}^2}{ L_\mathrm{Sun} d ^2} $$

**Task:** Since we know $d_\mathrm{Earth~to~Sun}=4.848 \times 10^{-6}$ pc and we have the distances to all our stars in the table, let's calculate and make a histogram of $\log_{10} (L / L_\mathrm{Sun})$. Use the $g$-band magnitude.

```python
distances = gaia['distance'] # pc
flux_relative_to_sun = magnitude_to_flux(gaia['gmag'])
L_relative_to_sun    = flux_relative_to_sun  *(gaia['distance'] / 4.848e-6)**2

plt.hist(np.log10(L_relative_to_sun))
plt.xlabel('$\log_{10}(L/L_\mathrm{sun})$')
```
